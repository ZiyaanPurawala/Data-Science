```python
fig,ax=plt.subplots(2,2)
fig ,axes=plt.subplots(2,2,layout='constrained',sharex=True,sharey=True)
ax[0,0].plot(j.index,j.DSA,"m-^")
ax[0,1].plot(j.index,j.DSA,"b*")
ax[1,0].plot(j.index,j.DSA,"rs:")
ax[1,1].plot(j.index,j.DSA,"b")
axes[1,1].plot(j.index.j.DSA,"b:")
axes[1,1].plot(j.index,j.DSA,"gs:")
```

```python
fig ,ax1=plt.subplots(2,2,layout='constrained',sharex=True,sharey=True)
```

```python
ax1[0,0].plot(j.index,j.DSA,"rs:")
ax1[0,0].legend("P")
ax1[0,1].plot(j.index,j.TESTING,"b<:","")
ax1[0,1].legend("T")
ax1[1,0].plot(j.index,j.NET,"g+-.",)
ax1[1,0].legend("D")
ax1[1,1].plot(j.index,j.DSA,"rs:")
ax1[1,1].plot(j.index,j.TESTING,"bs:")
ax1[1,1].plot(j.index,j.NET,"gs:")
ax1[1,1].legend("PDB")
ax1[0,0].set_title("DSA")
ax1[0,1].set_title("TESTING")
ax1[1,0].set_title("NET")
ax1[1,1].set_title("ALL SUBJECTS")
```

```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.displot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"],kde=False,bins=15)
```

# Jointplot
#jointplot() allows you to basically match up two displots for bivariate data.what kind parameters to compare with:

."scatter" ."reg" ."resid" ."kde" ."hex"

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="scatter")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="resid")
```

# pairplot
pairplot will plot pairwise relationship across an entire dataframe (for the numeric columns) and supports a color hue argyument (for categorical columns)

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue="sex",palette="brg")
```
