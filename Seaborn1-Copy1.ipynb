```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.displot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"],kde=False,bins=15)
```

# Jointplot
#jointplot() allows you to basically match up two displots for bivariate data.what kind parameters to compare with:

."scatter"
."reg"
."resid"
."kde"
."hex"

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="scatter")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="resid")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex")
```

# pairplot
pairplot will plot pairwise relationship across an entire dataframe
(for the numeric columns) and supports a color hue argyument (for categorical columns)

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue="sex",palette="brg")
```
